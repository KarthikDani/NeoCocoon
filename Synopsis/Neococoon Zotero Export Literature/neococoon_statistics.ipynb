```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data preparation
data = {
    'Category': ['Extremely Premature (<28 weeks)', 'Very Premature (28-32 weeks)', 'Moderate to Late Premature (32-37 weeks)'],
    'Percentage': [10, 10, 10],  # Adjust these values based on your needs
    'Deaths (2019)': [900000, 450000, 450000],  # Hypothetical numbers for demonstration
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the Seaborn style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='Category', y='Deaths (2019)', data=df, palette='pastel')

# Adding annotations for percentage
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                      textcoords='offset points')

# Title and labels
plt.title('Premature Birth Statistics', fontsize=16)
plt.xlabel('Prematurity Categories', fontsize=14)
plt.ylabel('Number of Deaths in 2019', fontsize=14)
plt.xticks(rotation=15)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data preparation for neonatal deaths in 2022
data = {
    'Category': [
        'Total Daily Newborn Deaths', 
        'Deaths in the First Week', 
        'Deaths after the First Week'
    ],
    'Number of Deaths': [
        6500,                # Daily newborn deaths
        6500 * 0.75,        # Deaths in the first week (75% of daily)
        6500 * (1 - 0.75)   # Deaths after the first week (25% of daily)
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the Seaborn style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='Category', y='Number of Deaths', data=df, palette='pastel')

# Adding annotations for number of deaths
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():,.0f}', 
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', 
                      fontsize=12, color='black', 
                      xytext=(0, 10), textcoords='offset points')

# Title and labels
plt.title('Neonatal Death Statistics for 2022', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Number of Deaths per Day', fontsize=14)
plt.xticks(rotation=15)
plt.tight_layout()

# Show the plot
plt.show()
```
